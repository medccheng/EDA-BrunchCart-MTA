# Brunch Cart in the NYC Financial District

The goal of this project is to help a client, who wants to start a mobile Brunch Cart around the NYC Financial District during weekdays, create a solid business plan. 

We can help us explore this goal by identifying which subway stations in the NYC Financial district have the highest traffic.

The figures below, display the Total Number of Turnstile Entries during weekdays from 8am to 4pm for each Subway Station in the Financial District. Charts are displayed for the months of June, July and August in 2021, respectively. 

These charts suggest that Chambers St, Bowling Green, and Wall St subway stations consistently have the highest foot traffic among all the stations located in the NYC Financial District for the past 3 months of 2021. However, both Rector St and World Trade Ctr stations should also be considered since the total entries for each is close to the Wall St's total entries.
